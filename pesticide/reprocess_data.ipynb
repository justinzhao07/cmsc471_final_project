```python
mast_data = pd.read_csv("./data/master_dataset.csv")
mast_data.head()
```

```python
list_top10 = ["Cereals, primary", "Sugar Crops Primary", "Sugar cane", "Hen eggs in shell, fresh", "Roots and Tubers, Total", "Vegetables Primary", "Maize (corn)", "Milk, Total", "Rice", "Wheat"]
mast_data = mast_data[mast_data["Item"].isin(list_top10)].reset_index(drop=True)
mast_data = mast_data[(mast_data["Year"] >= 1990) & (mast_data["Year"] <= 2022)].reset_index(drop=True)
mast_data["Area"] = mast_data["Area"].replace(country_map)
mast_data = mast_data[mast_data["Area"].isin(visible_countries["properties.name"])].reset_index(drop=True)
mast_data = mast_data.drop_duplicates(subset=['Area', 'Item', 'Year'], keep='first')
mast_data.head(11)
```

```python
group_cols = ['Area', 'Item']
numerical_cols = mast_data.select_dtypes(include='number').columns.difference(['Year'])

# 4. Group and compute percent change
def compute_percent_change(group):
    group = group.sort_values('Year')
    # Compute % change for each numerical column
    pct_change = group[numerical_cols].pct_change() * 100
    pct_change.insert(0, 'Year', group['Year'])  # Keep year for alignment
    return pct_change

# 5. Apply groupby
percent_change_df = mast_data.groupby(group_cols, group_keys=False).apply(compute_percent_change)

# 6. Reattach Country and Crop columns
percent_change_df[['Area', 'Item']] = mast_data[['Area', 'Item']].values

# 7. Optional: Rearrange columns
cols = ['Area', 'Item', 'Year'] + list(numerical_cols)
percent_change_df = percent_change_df[cols]

# 8. Sort for readability 
percent_change_df = percent_change_df.sort_values(by=['Area', 'Item', 'Year']).reset_index(drop=True)

percent_change_df.head(11)
```
