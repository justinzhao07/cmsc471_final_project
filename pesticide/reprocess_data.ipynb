```python
import pandas as pd
import requests
```

```python
data = pd.read_csv("./data/master_dataset.csv")
data = data.dropna(subset=["Pesticide_Use_per_Cropland"])
simple = data[["Area", "Year", "Pesticide_Use_per_Cropland"]].drop_duplicates().reset_index(drop=True)

simple.head()
```

```python
response = requests.get('https://raw.githubusercontent.com/holtzy/D3-graph-gallery/master/DATA/world.geojson')
vizible = response.json()

visible_countries = pd.json_normalize(vizible['features'])
visible_countries
```

```python
country_map = {
    'United States of America': 'USA',
    'Venezuela (Bolivarian Republic of)': 'Venezuela',
    'Bolivia (Plurinational State of)': 'Bolivia',
    'United Kingdom of Great Britain and Northern Ireland': 'England',
    'Netherlands (Kingdom of the)': 'Netherlands',
    'Czechia': 'Czech Republic',
    'Russian Federation': 'Russia',
    'Serbia': 'Republic of Serbia',
    'North Macedonia': 'Macedonia',
    'Republic of Moldova': 'Moldova',
    'Türkiye': 'Turkey',
    'Syrian Arab Republic': 'Syria',
    'Palestine': 'West Bank',
    'Guinea-Bissau': 'Guinea Bissau',
    "Côte d'Ivoire": 'Ivory Coast',
    'Congo': 'Republic of the Congo',
    'Eswatini': 'Swaziland',
    'Iran (Islamic Republic of)': 'Iran',
    'Republic of Korea': 'South Korea',
    'China, Taiwan Province of': 'Taiwan',
    "Lao People's Democratic Republic": 'Laos',
    'Viet Nam': 'Vietnam',
    'Bahamas': 'The Bahamas'
}
```

```python
simple["Area"] = simple["Area"].replace(country_map)
only_visible = simple[simple["Area"].isin(visible_countries["properties.name"])].reset_index(drop=True)
# only_visible["Area"].drop_duplicates().to_excel("boof.xlsx")
print(only_visible.shape[0])
only_visible.head(25)
```

```python
only_visible.rename(columns={"Area": "country", "Year": "year", "Pesticide_Use_per_Cropland": "pesticides_per_cropland"}, inplace=True)
only_visible.to_csv("./data/pesticides_clean.csv", index=False)
```
