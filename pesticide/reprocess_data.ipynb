```python
mast_data = pd.read_csv("./data/master_dataset.csv")
mast_data.head()
```

```python
list_top10 = ["Cereals, primary", "Sugar Crops Primary", "Sugar cane", "Hen eggs in shell, fresh", "Roots and Tubers, Total", "Vegetables Primary", "Maize (corn)", "Milk, Total", "Rice", "Wheat"]
mast_data = mast_data[mast_data["Item"].isin(list_top10)].reset_index(drop=True)
mast_data = mast_data[(mast_data["Year"] >= 1990) & (mast_data["Year"] <= 2022)].reset_index(drop=True)
mast_data["Area"] = mast_data["Area"].replace(country_map)
mast_data = mast_data[mast_data["Area"].isin(visible_countries["properties.name"])].reset_index(drop=True)
mast_data = mast_data.drop_duplicates(subset=['Area', 'Item', 'Year'], keep='first')
mast_data.head(11)
```

```python
prev_data = mast_data.copy()
prev_data["Year"] += 1

# Merge current data with previous-year data
merged = pd.merge(
    mast_data,
    prev_data,
    on=["Area", "Item", "Year"],
    suffixes=("", "_prev")
)

# Compute percentage change for each variable
def percent_change(current, previous):
    return (current - previous) / previous

merged["nitroChange"] = percent_change(
    merged["Cropland nitrogen per unit area"],
    merged["Cropland nitrogen per unit area_prev"]
)

merged["phosphorusChange"] = percent_change(
    merged["Cropland phosphorus per unit area"],
    merged["Cropland phosphorus per unit area_prev"]
)

merged["potassiumChange"] = percent_change(
    merged["Cropland potassium per unit area"],
    merged["Cropland potassium per unit area_prev"]
)

merged["pestChange"] = percent_change(
    merged["Pesticide_Use_per_Cropland"],
    merged["Pesticide_Use_per_Cropland_prev"]
)

merged["yieldChange"] = percent_change(
    merged["Production"],
    merged["Production_prev"]
)

# Merge the calculated changes back to the original dataset
mast_data = pd.merge(
    mast_data[["Area", "Item", "Year", "Temperature Change Compared to Previous Year"]],
    merged[["Area", "Item", "Year", "nitroChange", "phosphorusChange", "potassiumChange", "pestChange", "yieldChange"]],
    on=["Area", "Item", "Year"],
    how="left"
)

mast_data.rename(columns={"Temperature Change Compared to Previous Year": "tempChange"}, inplace=True)
mast_data.reset_index(drop=True, inplace=True)

```
